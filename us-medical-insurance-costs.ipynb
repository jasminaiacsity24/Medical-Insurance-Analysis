# U.S. Medical Insurance Costs

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("insurance.csv")

print("First 10 rows")
print(df.head(10))

print("Dataset info")
df.info()

print("Summary statistics")
print(df.describe())


def average_age(data):
    return data["age"].mean()

def average_charges(data):
    return data["charges"].mean()

def region_distribution(data):
    return data["region"].value_counts()

def smoker_vs_nonsmoker_cost(data):
    return data.groupby("smoker")["charges"].mean()

def avg_age_with_children(data):
    return data[data["children"] > 0]["age"].mean()

def bmi_charge_correlation(data):
    return data[["bmi", "charges"]].corr()

def age_charge_correlation(data):
    return data[["age", "charges"]].corr()

def children_vs_charges(data):
    return data.groupby("children")["charges"].mean()

print("Key Results")
print(f"1. Average age: {average_age(df):.2f} years")
print(f"2. Average medical charges: ${average_charges(df):,.2f}")

print("3. Region distribution")
print(region_distribution(df))

smoker_costs = smoker_vs_nonsmoker_cost(df)
print(f"4. Average charges for non-smokers: ${smoker_costs['no']:,.2f}")
print(f"5. Average charges for smokers: ${smoker_costs['yes']:,.2f}")

print(f"6. Average age (people with children): {avg_age_with_children(df):.2f} years")
print(f"7. Correlation between BMI and charges: {bmi_charge_correlation(df)}")
print(f"8. Correlation between age and charges: {age_charge_correlation(df)}")

print("9. Average charges by number of children")
print(children_vs_charges(df))



plt.scatter(df["bmi"], df["charges"])
plt.xlabel("BMI")
plt.ylabel("Medical Charges")
plt.title("BMI vs Medical Charges")
plt.show()

smokers = df[df["smoker"] == "yes"]
non_smokers = df[df["smoker"] == "no"]

plt.scatter(smokers["bmi"], smokers["charges"], color="red", label="Smokers")
plt.scatter(non_smokers["bmi"], non_smokers["charges"], color="purple", label="Non-Smokers")
plt.xlabel("BMI")
plt.ylabel("Medical Charges")
plt.title("BMI vs Charges (Smokers vs Non-Smokers)")
plt.legend()
plt.show()
```
